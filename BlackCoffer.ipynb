```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_json("/content/jsondata.json")
df.shape
```

```python
df=df.mask(df == '')      # Changed empty string to NaN value
df.head(5)
```

### **DATA ANALYSIS**

Pestle Analysis

```python
sb.countplot(data=df, x="pestle")
plt.title("Pestle Analysis")
plt.xticks(fontsize=10, rotation = 75)
plt.yticks(fontsize=14)
plt.show()
```

Most recorded sector: which sector was observed more

```python
#Most recorded sector

plt.figure(figsize=(16,4))
ax = sb.countplot(data = df, x = "sector")
for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.title("Most recorded sector")
plt.xticks(fontsize=10, rotation = 75)
plt.yticks(fontsize=14)
```

Sector vs Intensity: Determines the intensity of a sector at which they function

```python
df['intensity'] = pd.to_numeric(df['intensity'])
plt.figure(figsize=(22,6))
sb.boxplot(x = "sector", y = "intensity", data =df)
plt.title("sector vs intensity: Determines the intensity of a sector at which they function")
plt.xticks(fontsize=10, rotation = 75)
plt.yticks(fontsize=14)
```

region vs Intensity: Determines the intensity of a region at which they function

```python
plt.figure(figsize=(22,6))
sb.boxplot(x = "region", y = "intensity", data =df)
plt.title("region vs intensity: Determines the intensity of a region at which they function")
plt.xticks(fontsize=10, rotation = 75)
plt.yticks(fontsize=14)
```

country vs Intensity: Determines the intensity of a country at which they function

```python
plt.figure(figsize=(22,6))
sb.boxplot(x = "country", y = "intensity", data =df)
plt.title("country vs intensity: Determines the intensity of a country at which they function")
plt.xticks(fontsize=10, rotation = 90)
plt.yticks(fontsize=14)
```

Sector based analysis

```python
sb.set(style="ticks", color_codes=True)
g =sb.pairplot(df, hue ='sector')
g.fig.suptitle("sector analysis")
g.fig.set_size_inches(22,6)
plt.show()
```

Region based analysis

```python
df['region'] = df['region'].str.replace('world','World') # world and World

sb.set(style="ticks", color_codes=True)
g =sb.pairplot(df, hue ='region')
g.fig.suptitle("  region analysis")
g.fig.set_size_inches(22,6)
plt.show()
```

Country based analysis

```python
sb.set(style="ticks", color_codes=True)
g =sb.pairplot(df, hue ='country')
g.fig.suptitle("country analysis")
g.fig.set_size_inches(22,6)
plt.show()
```

Topic published count: which topic was observed more

```python
plt.figure(figsize=(32,6))
ax1 = sb.countplot(data= df, x = "topic")
for p in ax1.patches:
   ax1.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()+0.2), color='black', size=9, ha="center")
plt.title("Topic published count")
plt.xticks(fontsize=10, rotation = 90)
plt.yticks(fontsize=14)

plt.show()
```
